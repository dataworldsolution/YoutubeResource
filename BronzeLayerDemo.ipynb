```python
fl_path = 'abfss://HealthDataProject@onelake.dfs.fabric.microsoft.com/LH_Health.Lakehouse/Files/Landing/Hospital_Inpatient_Discharges_(SPARCS_De-Identified)__2016_20251109.csv'
```

```python
df = spark.read.format('csv').option('header','true').load(fl_path)
display(df)
```

```python
df.dtypes
df.printSchema()
```

```python
for col_name in df.columns:
    df = df.withColumnRenamed(col_name , col_name.replace(" ","_").replace("-","_"))
```

```python
df = df.withColumnRenamed('Zip_Code___3_digits',"Zip_Code_3_digits")
```

```python
df.write.mode('overwrite').format('delta').saveAsTable('LH_Health.Bronze.bronze_health_data')
```

```python
df = spark.sql("select count(*) from LH_Health.Bronze.bronze_health_data")
display(df)
```
